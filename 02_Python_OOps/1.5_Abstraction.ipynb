Abtstractin is the concept of hiding the complex implementation details adn showing only teh necessary features of an object . This helps in reducing programming complexity and efforts


```python
from abc import ABC,abstractmethod
## Abstract base class
class vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving ")
    
    @abstractmethod
    def start_engine(self):
        pass
class Car(vehicle):
    def start_engine(self):
        print("Car engine started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car=Car()
operate_vehicle(car)
```
