```python
## Basic methods
class person :
    def __init__(self,name,age):
        self.name=name
        self.age=age
p1=person("Ayush","20")
print(p1)
```

```python
## Basic methods
class person :
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def __str__(self):
        return f"{self.name},{self.age} years old"
    ## For our own custom representation of string
    def __repr__(self):
        return f"person(name={self.name},age={self.age})"

p1=person("Ayush","20")
print(p1)
print(repr(p1))
```
