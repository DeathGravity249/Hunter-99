```python
## Inheritance
## Parent class
class car:
    def __init__(self,windows,doors,engine_type):
        self.windows=windows
        self.doors=doors
        self.engine_type=engine_type
    def drive(self):
        print(f"The person will drive the {self.engine_type} car")

        
```

```python
car1 =car(4,5,"diesel_type")
car1.drive()
print(car1.drive()) ## if we want to use this we need to return the string above
```

```python
class tesla(car): ## Here first is child then in ()->parent class
    def __init__(self,windows,doors,engine_type,isself_selfdriving):
        super().__init__(windows,doors,engine_type)
        self.isself_selfdriving=isself_selfdriving
    def selfdriving(self):
        print(f"Tesls suport self driving  {self.isself_selfdriving}")

```

```python
tesla1=tesla(4,5,"electric",True)
tesla1.selfdriving()
```

```python
## Here i didnot wrote the drive fn in child classs
tesla1.drive()
```

```python
## Multiple Inheritance
## When a class inherit from more than one base class

class animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("Subclasses must implement this method")
## Base class 2
class pet:
    def __init__(self,owner):
        self.owner=owner

## Derived class
class Dog(animal,pet):
    def __init__(self, name,owner):
        animal.__init__(self,name)
        pet.__init__(self,owner)
    def speak(self):
        return f"{self.name}says woof"
    
        
## create an object 
dog1 =Dog("Milo","Ayush")
print( "The dog name is : ",dog1.name)
print("The dog owner : ",dog1.owner)
dog1.speak() ## parent class also contain the speak but priority is given to dog class
```
