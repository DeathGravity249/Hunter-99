```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
        
print_numbers(1,2,3,4,5,"Ayush Kumar")
```

```python
## Keywords arguments 
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
print_details(name="Ayush",age="32",contry="India")
```

```python
## Using both the argument
def print_details(*args,**kwargs):
    for value in args:
        print(f"Positional arguments : {value}")
    for key,value in kwargs:
        print(f"{key}:{value}")
print_details(1,2,3,4,name="Ayush Kumar",age=32,Grade="S")
```
