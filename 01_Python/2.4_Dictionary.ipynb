```python
## Iterating over dictionaries
## You can use loopsto iterate over dictionaries , keys,values, or items
## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterating over values
for value in student.values():
    print(value)
```

```python
# Iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested  Dictionary 
students={
      "student1":{"Name":"Ayush","age":20,"Grade":"A"},
      "student2":{"Name":"shresth","age":21,"Grade":"S"}
   }
print(students)
```

```python
## Access nested dictionaries elements
print(students["student1"]["Name"])
print(students["student2"]["age"])
## It is somewhat similar to nested loop
```

```python
students.items()
```

```python
## Iterating over nested dicitonaries 
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")

    
```

```python
## Dictionary comprehension
## This one is for remembering normal fn
squares={x**2 for x in range(5)}
print(squares)
## This for condition -> key: value
squares={x:x**2 for x in range(5)}
print(squares)

```

```python
## Conditional list comprehension
sq={x:x**2 for x in range(7) if x%2==0}
print(sq)
```

```python
## Practical examples
## To count frequency of element
numbers=[1,2,3,3,2,4,4,5,6,]
frequency={}
for number in numbers:
    if number in frequency :
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```
