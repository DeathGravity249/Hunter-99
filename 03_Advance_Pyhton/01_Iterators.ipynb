Iterators are advance Python concepts that allow for efficient looping and memory mangagement . Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure

```python
my_list=[1,2,3,4,5]
for i in my_list:
    print(i)
```

```python
type(my_list)
print(my_list)
```

```python
# Iterator
iterator=iter(my_list)
print(type(iterator))
```

```python
iterator
```

```python
## Above will not get any element unless we iterate it 
next(iterator)
```

```python
iterator=iter(my_list)
```

```python
## First it will print 
# then it will stop iterating when no. end
try:
    print(next(iterator))

except StopIteration:
    print("There are no element stop the iteration")
```

```python
# Here it get redefined everytime 
my_str="Hello"
string_iterator=iter(my_str)

print(next(string_iterator))
print(next(string_iterator))
```
