# Syntax of map
map(function, iterable)


```python
## Using lambda and map
numbers=[11,22,33,44,55]
square=map(lambda x :x*x,numbers )
print(list(square))
```

✅ When to Use map()
When you want to transform all elements in a sequence.

When you want cleaner and more functional code (compared to loops).

Often combined with lambda for concise syntax.

```python
## Map with multiple iterable
numbers1=[12,34,55,66,7]
numbers2=[11,22,33,44,55]
add_square=map(lambda x,y:x+y,numbers1,numbers2)
print(list(add_square))
```

```python
## map() to convert a list of string to intergers
# Use map to convert strings to integers
str_numbers=['1','2','3','4']
int_numbers=list(map(int,str_numbers)) # typecast
print(int_numbers)
```

```python
words=['apple','banana','cherry']
upper_world=list(map(str.upper,words))
print(upper_world)
```

```python
def get_name(person):
    return person['name']
people=[
    {'name':'krish','age':32},
    {'name':'Ayush','age':16}
]
list(map(get_name,people))

```
