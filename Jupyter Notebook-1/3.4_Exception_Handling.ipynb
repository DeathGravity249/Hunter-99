```python
## reference code hwo things work
try:
    # code that may raise different exceptions
    result = 1 / 0
    a = b
    my_list = [1, 2, 3]
    print(my_list[5])
except ZeroDivisionError:# IF any block runs code is stopped
    print("Cannot divide by zero!")
except NameError:
    print("Variable not defined!")
except IndexError:
    print("Index out of range!")
except Exception: ## This runs only when none above runs 
    print("Some other error occurred!")
```

```python
try:
    num=int(input("Enter a number "))
    result=10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("Enter the denominator greator than 0")
except Exception as ex:
    print(ex)
```

```python
## try,except ,else block 
try:
    num =int(input("Enter the input"))
    result=num/10
except ValueError:

    print("That's not a valid number ")
except ZeroDivisionError:
    print("You can't divide by zero!")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}")
```

```python
## try,except ,else and finally
try:
    num =int(input("Enter the input"))
    result=num/10
except ValueError:

    print("That's not a valid number ")
except ZeroDivisionError:
    print("You can't divide by zero!")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}")
finally:## Cleanup code
    print("Execution is comlete ")
```

```python
## File handling an Exception Handling
try:
    file=open('example1.txt','r')
    content=file.read()
    print(content)
except FileNotFoundError:
    print("The file is not present ")
finally:
    if 'file' in locals() and not file.closed():
        file.close()
        print('file close')

```
