```python
# Create a set 
my_set={1,2,3,4,5,6}
print(my_set)
print(type(my_set))
```

```python
my_set=set([1,2,3,4,5,60])
print(type(my_set))
```

```python
# Now type duplicate the element 
my_set={1,2,3,4,5,6,5,2}
print(my_set)
```

```python
## Basic set operation 
## Adding and removing 
my_set={1,2,3,4,5,6,5,2}
my_set.add(7)
print(my_set)
my_set.add(6)
print(my_set)
## Now to remove
print("Now use remove fun: ")
my_set.remove(3)
## Now using the discard fn
my_set.discard(11) ## this will not show the error as ->remove if 
print(my_set)
```

```python
## Pop method
my_st={9,8,7,6,5,4,3}
remove_element=my_st.pop() ## search for smallest value 
## elements are not ordered or have index like list
print(my_st)
```

```python
## clear all element
my_st.clear()
print(my_st)
```

```python
## set membership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## mathematical operation 
set1={1,2,3,4,5}
set2={4,5,6,7,8,9}

## Union
union_set=set1.union(set2)
print(union_set)

## Intersection
intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5}
set2={4,5,6,7,8,9}
## Difference 
print(set1.difference(set2))
print(set2.difference(set1))
```

```python
## Symmeteric Difference
## The following function combine unique element from each element
set1.symmetric_difference(set2)
```

```python
## Sets methods
set1={1,2,3}
set2={3,4,5}

## is subset
print(set1.issubset(set2))
set2={1,2,3,4,5}
print(set1.issubset(set2))
print(set2.issuperset(set1))
```

```python
## Counting unique word in text
text="In this tutorial we are discussing about sets "
words=text.split()

## Convert List of words to set to get unique words

unique=set(words)
print(unique)
print(len(unique))
```
