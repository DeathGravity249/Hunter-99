### List comprehension
Basic syntax [expression for item in iterable]

with conditional logic : 
[Expression for item in iterable if condition]

```python
### Basic comprehension
square = [num**2 for num in range(10)]
print(square)
```

```python
##3 List comprehension with condition

list=[]
for i in range(10):
    if(i%2==0):
        list.append(i)
print(list)
 
```

```python
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```

#### List Comprehension
Basic syantax           [expression for item in iterable]

with conditional logic   [expresssion for item in iterable if condition ]
 
Nested List Comprehension  [expression for item1 in iterable for item2 in interable]

```python
## Nested list comprehension
lst1 = [1,2,3,4]
lst2=['a','b','c','d','e']

pair=[[i,j] for i in lst1 for j in lst2]
print(pair)
```

```python
## List comprehension with funcion call
fruits=["Apple ", "Bannna ","Gauva", "Litchi","kiwi"]
fruit=[len(fruit) for fruit in fruits]
# Happening here -> loop iterating 
print(fruit)
```
