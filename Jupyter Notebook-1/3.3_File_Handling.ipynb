```python
## Write a file (without overwriting)
with open('example.txt','a') as file:
    file.write("Append operation taking place!")
```

```python
## Writing a list fo lines to a file
lines=['\nFirst line \n','second line \n','Third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## binary files
# writing to a binary file
data =b'\x00\x01\x02\x03'
with open ('example.bin','wb')as file:
    file.write(data)

```

```python
## Reading the bin file
with open('example.bin','rb') as file:
    content =file.read()
    print(content)
```

```python
## Read teh content from a source text file adn write to a destination text file
## Copying a text file
with open('example.txt','r') as sourc_file:
    content=sourc_file.read()
with open('destination.txt','w') as destination_file :
    destination_file.write(content)
```

The w+ mode in Python is used to open a file for both reading and writing . If the file does not exist , it will be created .If the file exists , its content is truncated(i,e, the file iis overwritten)

```python
## Writing and then reading a file
with open ('example.txt','w+')as file:
    file.write("Hello world\n")
    file.write("This is new line \n")

# Move the file cursor to the begining 
    file.seek(0)

## Read the content of the file
    content=file.read()
    print(content)
```
