```python
## Doing stuff with normally 
def even(num):
    if num%2==0:
        return True
even(24)
```

```python
# Now using the lambda function
# Arguments need to passed 
even1=lambda num:num%2==0
even1(12)
```

```python
addition = lambda x,y,z: x+y+z
print("Output is : ", addition(12,13,14))
```
