```python
## WHILE LOOP 
count=0
while count<5:
    print(count)
    count=count+1
```

```python
# loop with the usage of break statement

for i in range(1,7):
    if(i==5):
        break
    print(i)
```

```python
# continue -> skip the current statment and continues with next
for i in range(10):
    if(i%2==0):
        continue
    print(i)
```

```python
## PASS -> THE PASS STATEMENT IS NULL OPERATION ; IT DOES NOTHING 
for i in range(5):
    if(i%2==0):
        pass
    print(i)
```

```python
# Nested loops inside loops
for i in range(3):
    for j in range(2):
        print(f"i is : {i} and j is {j}")
```

```python
## First n natural numbers
x=int(input("Enter the number upto which recquired : "))
sum =0
while(x!=0):
    sum=sum+x
    x-=1
print("The result is : ",sum)
```

```python
## Example Prime number between 1 to Hunderd
for num in range(1,10):
    if num >1:
        for i in range(2,num):
           if num%i==0:
            break
        else:
            print(num)
```
