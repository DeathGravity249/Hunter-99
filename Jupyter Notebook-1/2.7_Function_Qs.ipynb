Example 1 : Temperature conversion

```python
def convert_temperature(temp, unit):
    """This function converts the temperature."""
    if unit.lower() in ("c", "celsius"):
        return temp * (9/5) + 32
    elif unit.lower() in ("f", "fahrenheit"):
        return (temp - 32) * 5/9
    else:
        return "Invalid input"

x = int(input("Enter the temperature: "))
y = input("Enter the unit (c/celsius or f/fahrenheit): ")

print(convert_temperature(x, y))
```

Example 3: Calculate the total cost of Items in a shopping cart

```python
def calculate_total_cost(cart):
    total_cost=0
    for item in cart:
        total_cost+=item['price']*item['quantity']
    return total_cost

## Example cart data
cart=[
    {'name':'Apple','price':0.5,'quantity':4},
    {'name':'Litchi','price':0.9,'quantity':9},
    {'name':'Guava','price':0.87,'quantity':6}
]
## calling the function
total_cost=calculate_total_cost(cart)
print(total_cost)
```

Example 4 : Check if a string is palindrome 

```python
def is_palindrome(s):
    s=s.replace(" ","").lower()
    if s[0:]==s[::-1]:
        print("The given input is palindrome")
    else :
        print("The given input is not palindrome")
str=input("ENter the string :")
is_palindrome(str)
```

Exampl 5:
Calculate the factorial of a number using recursion


```python
def factorial(n):
    if n==0:
        return 1
    else :
        return n*factorial(n-1)
print("The output is : ",factorial(5))
```

 Example 6: Function to read a file and count the frequency of each word

```python
# The function’s job is to read the file, 
# process the text, and return a dictionary 
# containing the frequency of each word.
def count_word_frequency(file_path):
    word_count={}
    ## with ensure no harm occur open safely ,close safely
    with open(file_path,'r') as file: 
        for line in file: ## iterate over each line
            words=line.split() ## split line in list of string 
            ## identified by the whitespace , tab etc
            for word in words:
                word=word.lower().strip('.,!?:"\';')
                word_count[word]=word_count.get(word,0)+1

    return word_count

filepath='sample.txt'
word_frequency=count_word_frequency(filepath)
print(word_frequency)
print(len(word_frequency))
```
